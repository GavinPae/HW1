```python
import pandas as pd
#data frame
data={
    'Name':['Alice','Bob','Charles', 'Dave'],
    'Age':[25,30,17,28],
    'City':['New York','San Francisco','Los Angeles','Alexandria'],
    'Salary':[100,150,200,300]
}
df=pd.DataFrame(data)
df
```

```python
df[['Name','Salary']]
df.iloc[1]
```

```python
#calculate column stats
df['Age'].mean()
```

```python
print("The minimum salary is",df['Salary'].min())
print("The maximum salary is",df['Salary'].max())
print("The mean salary is",df['Salary'].mean())
```

```python
ndata=[10,20,30,40,50]
s=pd.Series(ndata)
print(s.min())
print(s.max())
print(s.mean())
```

**Matplotlib**

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[10,15,7,12,8]
plt.plot(x,y, marker='+',markersize=8, color='blue', linestyle='-.', linewidth=2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('X versus Y')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Year': [2010,2011,2012,2013,2014],
    'Sales':[500,800,1700,1500,2000]
}
df=pd.DataFrame(data)

plt.plot(df['Year'],df['Sales'],marker='o')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Over Years')
plt.grid(True)
plt.xticks(df['Year'])
plt.show()
```

```python
data={
    'Category':['A','B','C','D'],
    'Sales':[500,800,1200,1500]
}
df=pd.DataFrame(data)
plt.bar(df['Category'],df['Sales'], color='red')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales by Category')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Age': [25,30,22,35,28],
    'Income':[5000,60000,4500,70000,55000]
}
df=pd.DataFrame(data)

plt.plot(df['Age'],df['Income'],marker='o')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Income vs. Age')
plt.grid(True)
plt.show()
```

```python
#creating a data frame
data={
    'Scores':[85,78,92,65,80,75,88,90,82,70]
}
df=pd.DataFrame(data)

plt.hist(df['Scores'], width = 1.0)
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('distribution of Scores')
plt.grid(True)
plt.show()
```
